```python
# Check for missing values in agency , count missing values,total missing in entire df
missing_values = agen.isnull()
print(missing_values)
missing_count = agen.isnull().sum()
print(missing_count)
```

```python
# Check for missing values in calendar , count missing values,total missing in entire df, rows and columns
missing_values = cal.isnull()
print('missing vals: \n', missing_values)
missing_count = cal.isnull().sum()
print('missing count: \n' , missing_count)
total_missing = cal.isnull().sum().sum()
print('Total missing values:\n', {total_missing})
rows_with_missing = cal[cal.isnull().any(axis=1)]
print('rows with misssing values:\n ' , rows_with_missing)
columns_with_missing = cal.columns[cal.isnull().any()]
print(f'Columns with missing values: {list(columns_with_missing)}\n')
```
